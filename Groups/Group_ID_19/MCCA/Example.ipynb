```python
X_list =[X1,X2,X3]
mcca = MCCA(n_components=3,reg_param=0.001)
mcca.fit([X1,X2,X3])
X_reduced = mcca.transform([X1,X2,X3])
for i in range(len(X_reduced)):
    print("X_reduced_{0}->{1}\n".format(i+1,X_reduced[i]))
  
```
