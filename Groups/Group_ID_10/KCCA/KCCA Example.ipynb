```python
#importing package
from KCCA import KCCA
import numpy as np
import pandas as pd
```

```python
#make Object by calling KCCA
kcca=KCCA()
```

```python
X = pd.read_csv('./mfeat-mor.csv',delim_whitespace=True,header=None)
Y = pd.read_csv('./mfeat-fou.csv',delim_whitespace=True,header=None)
```

```python
print(X.shape,Y.shape)
```

```python
#fitting data
wx,wy=kcca.fit(X,Y)

print("Projections: ")
print("Wx :: ",wx)
print("Wy :: ",wy)
```

```python
#Transformed matrix
Xnew,Ynew= kcca.fit_transform(X,Y)

print("Output transformed Matrix: ")
print("Xnew :: ",Xnew)
print("Ynew ::",Ynew)
```
