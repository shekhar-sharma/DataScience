```python
#import libraries
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
from skimage.io import imread,imshow
from skimage.filters import prewitt_h,prewitt_v
%matplotlib inline
```

```python
#read an image
image = imread(r"C:\Users\Lenovo.LAPTOP-O89TDI3L\OneDrive\Desktop\1. Project_Data_Science\image1.jpg", as_gray=True)
imshow(image)
```

```python
#shape of image
image.shape, image
```

```python
#detecting edges of image
horizontal_edges = prewitt_h(image)
vertical_edges = prewitt_v(image)

#vertical edges
imshow(vertical_edges, cmap='gray')
```

```python
#horizantal edges
imshow(horizontal_edges, cmap='gray')
```
