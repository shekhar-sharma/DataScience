W below is an diagonal matrices of size(n*n)

```python
row,col=(n,n)
W=[]
for i in range(col):
    c=[]
    for j in range(row):
        if i==j:
            c.append(1/n)
        else:
            c.append(0)
    W.append(c)
W
```

```python
row,col=(n,n)
I=[]
for i in range(col):
    c1=[]
    for j in range(row):
        if i==j:
            c1.append(1)
        else:
            c1.append(0)
    I.append(c1)
I
```

I is identity matrix of size(n*n) made to perform subtraction of 2 matrices (I-W)

```python
Diff=np.subtract(I,W)
Diff
```
